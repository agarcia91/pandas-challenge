```python
import pandas as pd
#import csv
school_data = '/Users/anthonygarcia/Desktop/PyCitySchools/schools_complete.csv'
student_data = '/Users/anthonygarcia/Desktop/PyCitySchools/students_complete.csv'
#read csv
school_info = pd.read_csv(school_data)
student_info = pd.read_csv(student_data)
#merge csv
school_data_complete = pd.merge(student_info, school_info, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#schools in district 
unique = school_data_complete["school_name"].unique()
unique
```

```python
#student total
student_count = school_data_complete["student_name"].count()
student_count
```

```python
#each schools budget
budget = school_data_complete["budget"].unique()
budget
```

```python
#total budget
total_budget = budget.sum()
total_budget
```

```python
#calculations of passing reading
reading_pass = school_data_complete.loc[school_data_complete["reading_score"] >=70]

student_passing_reading = reading_pass["Student ID"].count()

percent_passing_reading = (student_passing_reading/student_count)*100
percent_passing_reading

round_percent_reading_pass = round(85.80546336482001,6)
round_percent_reading_pass
```

```python
#group by school name
school_name = school_data_complete.set_index('school_name').groupby(['school_name'])

#total
school_types = school_info.set_index('school_name')['type']
total_students = school_name["Student ID"].count()
total_budget = school_info.set_index('school_name')['budget']

#Per Student Budget
per_student_budget = total_budget/total_students

#calculations
avg_math = school_name["math_score"].mean()
avg_reading = school_name["reading_score"].mean()
percent_math = school_data_complete.loc[school_data_complete["math_score"] >= 70].groupby('school_name')['Student ID'].count()/total_students*100
percent_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70].groupby('school_name')['Student ID'].count()/total_students*100
overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/total_students 
```

```python
#new df
school_summary = pd.DataFrame({"School Type" : school_types,"Total Students" : total_students,"Total School Budget" : total_budget,"Per Student Budget" : per_student_budget,"Average Math Score" : avg_math,"Average Reading Score" : avg_reading,"% Passing Math": percent_math,"% Passing Reading" : percent_reading,"Overall Passing Rate" : overall})
#formatting
school_summary['Overall Passing Rate'] = school_summary['Overall Passing Rate'].map("{:.6%}".format)
school_summary
```

```python
# top 5 schools
top5 = school_summary.sort_values('Overall Passing Rate', ascending=False)
top5.head(5)
```

```python
bins = [0, 584, 629, 644, 675]
group_name = ["<$584", "$585-629", "$630-644", "$645-675"]

school_data_complete['spending_bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], bins, labels = group_name)

#group by spending
_spending = school_data_complete.groupby('spending_bins')

#calculations, calculations are slightly off
avg_math = _spending['math_score'].mean()
avg_read = _spending['reading_score'].mean()
pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('spending_bins')['Student ID'].count()/_spending['Student ID'].count()*100
pass_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('spending_bins')['Student ID'].count()/_spending['Student ID'].count()*100
overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('spending_bins')['Student ID'].count()/_spending['Student ID'].count()*100
    
```

```python
#new df
spend_scores = pd.DataFrame({'Average Math Score':  avg_math, 'Average Reading Score': avg_read, '% Passing Math': pass_math, '% Passing Reading': pass_read, '% Overall Passing': overall}) 
spend_scores


#formatting
spend_scores['Average Math Score'] = spend_scores['Average Math Score'].map("{:.2f}".format)
spend_scores['Average Reading Score'] = spend_scores['Average Reading Score'].map("{:.2f}".format)
spend_scores['% Passing Math'] = spend_scores['% Passing Math'].map("{:.2f}".format)
spend_scores['% Passing Reading'] = spend_scores['% Passing Reading'].map("{:.2f}".format)
spend_scores['% Overall Passing'] = spend_scores['% Overall Passing'].map("{:.2f}".format)
spend_scores.index.name = "Spending Ranges (Per Student)"
spend_scores
```

```python
# create bins
bins = [0, 1000, 1999,5000]
group_name = ["Small (<1000)", "Medium (1000-2000)" , "Large (2000-5000)"]
school_data_complete['size_bin'] = pd.cut(school_data_complete['size'], bins, labels = group_name)

#group by spending
by_size = school_data_complete.groupby('size_bin')

#calculations 
avg_math = by_size['math_score'].mean()
avg_reading = by_size['math_score'].mean()
pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('size_bin')['Student ID'].count()/by_size['Student ID'].count()*100
pass_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('size_bin')['Student ID'].count()/by_size['Student ID'].count()*100
overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('size_bin')['Student ID'].count()/by_size['Student ID'].count()*100


#new dataframe
size = pd.DataFrame({ "Average Math Score": avg_math,"Average Reading Score": avg_reading,'% Passing Math': pass_math,'% Passing Reading': pass_read,'% Overall Passing': overall})
size

#give index a name
size.index.name = "Total Students"
size
```
